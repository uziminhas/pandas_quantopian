```python
returns = pd.DataFrame(np.random.normal(1.0,0.03,(100,10))) # 100 rows and 10 columns
prices = returns.cumprod() # cumulative product of returns
prices.plot()
plt.title('Randomly-generated Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc=0)
```

```python
s = pd.Series([1, 2, np.nan, 4, 5]) 
print(s) # will be automatically indexed
```

```python
print(s.name) # every series has a name
```

```python
s.name = 'Toy Series'
print(s.name)
```

```python
print(s.index)
```

```python
# declaring a new index variable
new_index = pd.date_range("2018-01-01", periods=len(s), freq="D")
print(new_index)
```

```python
s.index = new_index # setting declared index to this series' index
print(s.index)
```

```python
print("First element of the series: ", s.iloc[0])
print("Last element of the series: ", s.iloc[len(s) - 1])
```

```python
s.iloc[:2]
```

```python
start = 0
end = len(s) - 1
step = 1

s.iloc[start:end:step]
```

```python
# reverse a series by declaring a negative step
s.iloc[::-1]
```
